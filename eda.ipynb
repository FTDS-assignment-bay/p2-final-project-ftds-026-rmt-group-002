```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df = pd.read_csv('data_clean_final.csv')

df
```

```python
# sort tabel berdasarkan date_of_purchase 
df = df.sort_values(by='date_of_purchase')
df
```

```python
df.info()
```

**Insight** :
Terdapat 14 kolom dengan mayoritas tipe data numerik, kecuali pada kolom product_category, payment_method, customer_name, date_of_purchase, dan time_of_purchase yang bertipe data object. dataframe memiliki jumlah row 12922 dengan 0 missing value

```python
# persentase churn
churn_rate = df['churn'].value_counts()
plt.pie(churn_rate.values, labels=churn_rate.index, autopct='%1.1f%%')
plt.title('Churn Rate')
plt.show()
```

**Insight** : sepanjang 2020-2023, terjadi *churn rate* dengan persentase sebanyak 50% customer


```python
# pengaruh return terhadap churn rate

# Konversi tipe data kolom 'churn' ke string
df['churn'] = df['churn'].astype(str)

# plot bar churn by return
sns.countplot(x='returns', hue='churn', data=df)
plt.title('Churn by Return')
plt.xlabel('Return')
plt.ylabel('Count')
plt.show()
```

**Insight** : berdasarkan return atau customer yang mengembalikan barang, menunjukkan bahwa return/tidak return memiliki rasio yang mirip terhadap *churn*

```python
# persentase berdasarkan gender

gender_rate = df['gender'].value_counts()
plt.pie(gender_rate.values, labels=gender_rate.index, autopct='%1.1f%%')
plt.title('Gender Rate')
plt.show()
```

**Insight** : persentase customer tidak terlalu berbeda antar gender, customer wanita memiliki persentase sedikit lebih banyak dari customer pria  **namun tidak signifikan**

```python
# pengaruh return terhadap churn rate

ax = sns.countplot(x='gender', hue='churn', data=df)
for p in ax.patches:
    height = p.get_height()
    width = p.get_width()
    x, y = p.get_xy() 
    ax.annotate(f'{height}', (x + width/2, y + height), ha='center')
plt.title('Churn by Gender')
plt.xlabel('gender')
plt.ylabel('Count')
plt.show()
```

**Insight** : dapat dikatakan bahwa perbedaan gender tidak mempengaruhi churn rate. Terdapat perbedaan churn berdasarkan gender wanita dan pria, dimana sedikiy lebih banyak jumlah customer pria yang mengalami churn  **namun tidak signifikan**

```python
# distribusi age

sns.histplot(x='age', data=df)
plt.title('Histogram Age')
plt.xlabel('Age')
plt.ylabel('Freq')
plt.show()
```

**Insight** : pada umur customer pengguna *e-commerce*, terdapat lonjakan frekuensi pada umur 20, sekitar 25, sekitar 35, sekitar 50 an , sekitar 60an dan 70. Namun secara keseluruhan memiliki distribusi yang normal dan tidak condong atau didominasi dari umur tertentu

```python

# groupby antara product category dengan total purchase amount
purchase_amount_per_category = df.groupby('product_category')['total_purchase_amount'].size().sort_values(ascending=False).reset_index()

# bar plot jumlah frekuensi pembelian disetiap product category
ax = sns.barplot(x='product_category', y='total_purchase_amount', data=purchase_amount_per_category)
# menambahkan anotasi
for p in ax.patches:
    height = p.get_height()
    width = p.get_width()
    x, y = p.get_xy() 
    ax.annotate(f'{height}', (x + width/2, y + height), ha='center')
plt.title('Product Category')
plt.xlabel('Product Category')
plt.ylabel('Count')
plt.show()
```

**Insight** :
- Pada product category, hasil penjualan di e-commerce dari tahun 2021-2023 menunjukkan bahwa frekuensi penjualan terbanyak merupakan penjualan barang dengan kategori clothing/pakaian, disusul dengan kategori buku. kemudian kategori elektronik dan perabotan rumah yang memiliki frekuensi penjualan lebih rendah dibanding kategori pakaian dan buku.
- Terdapat kecenderungan bahwa customer menggunakan layanan e-commerce ini untuk produk fashion dan membeli buku dibanding untuk membeli peralatan elektronik maupun peralatan rumah tangga. Dapat diasumsikan kemungkinan customer lebih memilih untuk melakukan pembelian secara langsung ke toko fisik jika ingin membeli barang-barang yang bersifat elektronik maupun barang-barang untuk rumah, atau customer lebih memilih e-commerce lain untuk kategori produk barang elektronik dan peralatan rumah.

```python
# persentase payment method
payment_method_rate = df['payment_method'].value_counts()
plt.pie(payment_method_rate.values, labels=payment_method_rate.index, autopct='%1.1f%%')
plt.title('Payment Method')
plt.show()
```

**Insight** : 
- Pada payment method, distribusi tidak terlalu seimbang antar pilihan metode pembayaran customer, pembayaran customer untuk pembelian produk di e-commerce masih didominasi oleh credit card dengan persentase 40,3%, kemudian disusul dengan pembayaran paypal sebesar 30,3% kemudian pembayaran cash serta crypto sebagai metode pembayaran dengan persentase terkecil.
- Customer lebih senang menggunakan credit card dan paypal untuk melakukan pembelian barang di e-commerce. Hal ini dapat diakibatkan kemudahan terkait integrasi antara credit card dan paypal ke sistem e-commerce, serta kepraktisan dalam metode pembayaran tersebut dibandingkan dengan penggunaan cash maupun crypto dimana pengguna/pemilik crypto tentunya tidak terlalu banyak dibanding pengguna umum. Customer lebih familiar dengan kehadiran dan penggunaan credit card dan paypal untuk melakukan transaksi, termasuk dalam transaksi online e-commerce


```python
# pengaruh payment method terhadap churn rate

sns.countplot(x='payment_method', hue='churn', data=df)
plt.title('Churn by Payment Method')
plt.xlabel('Payment Method')
plt.ylabel('Count')
plt.show()
```

**Insight** : Payment method yang digunakan customer tidak berpengaruh terhadap kemungkinan *churn*, pada setiap payment method menunjukkan bahwa kemungkinan *churn* atau tidak *churn* memiliki persentase churn yang mirip dan tidak didominasi salah satu. Pada payment method credit card terlihat customer yang churn **sedikit lebih banyak** dari customer yang tidak *churn*, namun jumlahnya tidak terlalu signifikan.
